```python
import pandas as pd
import os
# import xlrd

# Define the directory path where the files are located
directory = 'dataset'

# Initialize an empty DataFrame to store the merged data
merged_data = pd.DataFrame()
data = pd.DataFrame()
i = 0 
# Iterate over each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.xlsx'):  # Assuming the files are in CSV format
        file_path = os.path.join(directory, filename)
#         print( file_path )
        # Read the file into a DataFrame
        data = pd.read_excel(file_path ,header=None  )

        # Delete the 40th row

        data.drop(2, inplace=True)
        data.drop(1, inplace=True)
        data.drop(0, inplace=True)
        data.drop(39, inplace=True)
#         print(data.head(41))
        
        no_of_col = data.shape[1]
        
        
        if(no_of_col == 16):
            data = data.drop(data.columns[[7, 8, 14 ]],axis = 1)

            
        if(no_of_col == 15):
            data = data.drop(data.columns[[7,13]],axis = 1)
            
        if(no_of_col == 14 ):
            data = data.drop(data.columns[[7]],axis = 1)

        
        data.reset_index()
        
        #Extracting Date from filename
        date_part = "-".join(os.path.splitext(filename)[0].split("-")[-3:])
        data[13] = str(date_part)
#         if(filename == "Corona-Fallmeldungen-RLP-2021-11-24.xlsx"):
#             print(data.head())
#             break

        
        
#         # Append the data to the merged_data DataFrame
#         i = i +1
#         print(i , " ",data.shape[1] , "  ", filename )
        merged_data = pd.concat([merged_data , pd.DataFrame(data)], ignore_index=True)
    
    
        #Cleaning Merged Dataset

        # Define the values to be dropped
        drop_values = ['Laborbestätigt, seit Beginn der Pandemie', 'Gesamt', 'vorherigen Berichtstag','Inzidenz der letzten 7-Tage pro 100.000C','Differenz zum' ]

        # Drop rows that contain the specified values in any cell
        merged_data = merged_data[~merged_data.isin(drop_values)].dropna(how='all')
        merged_data.reset_index(drop=True, inplace=True)



        
        #Dropping Empty Rows
        merged_data = merged_data.dropna(how='all')
        merged_data.reset_index(drop=True, inplace=True)
        
        #Taking only first 13 Columns
        merged_data = merged_data.iloc[:, :13]



#Renaming the columns
# Define the new column names
new_columns = {
    0: 'District',
    1: 'Laboratory confirmed, since the beginning of the pandemic TOTAL',
    2:'Laboratory confirmed, since the beginning of the pandemic Difference from',
    3: 'Laboratory confirmed, since the beginning of the pandemic Hospitalized',
    4: 'Laboratory confirmed, since the beginning of the pandemic Deceased',
    5: 'Laboratory confirmed, since the beginning of the pandemic RecoveredA',
    6: 'Laboratory confirmed, since the beginning of the pandemic Current CasesB',
    7: 'Reported, the last 7 days,Incidence per 100, Incidence per 100 ,Rheinland-Pfalz',
    8: 'Reported, the last 7 days,Incidence per 100, Incidence per 100 ,+USAFD',   
    9: 'Reported, the last 7 days,Incidence per 100, Incidence per 100 ,<20 Jahre',
    10: 'Reported, the last 7 days,Incidence per 100, Incidence per 100 ,20-59 JahreE',
    11: 'Reported, the last 7 days,Incidence per 100, Incidence per 100 ,>=60 Jahre',
    12: 'Date'
    # Add more index-name mappings as needed
}

# Rename the columns using index positions
merged_data.columns = [new_columns.get(i, col) for i, col in enumerate(merged_data.columns)]


# Save the merged data to a new CSV file
merged_data.to_excel('merged_data3_timeseries.xlsx', index=False )
# merged_data.to_csv('merged_data.csv', index=False )
merged_data.to_excel('merged_data4.xlsx', index=False )

merged_data.head()
```
